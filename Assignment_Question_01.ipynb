```python
# Madhumal Motors CLI-Based Inventory and Sales Management System 01

# Inventory List
inventory = []

# Function to add a motorcycle
def add_motorcycle():
    make = input("Enter make: ")
    model = input("Enter model: ")
    quantity = int(input("Enter quantity: "))
    price = float(input("Enter price: "))

    for bike in inventory:
        if bike['make'] == make and bike['model'] == model:
            bike['quantity'] += quantity
            print("Updated existing bike quantity.")
            return

    inventory.append({
        'make': make,
        'model': model,
        'quantity': quantity,
        'price': price
    })
    print("Motorcycle added successfully.")

# Function to search for a motorcycle price
def search_price():
    make = input("Enter make: ")
    model = input("Enter model: ")

    for bike in inventory:
        if bike['make'] == make and bike['model'] == model:
            print(f"The price of {make} {model} is LKR {bike['price']}")
            return
    print("Motorcycle not found.")

# Function to sell motorcycle and generate invoice
def sell_motorcycle():
    make = input("Enter make: ")
    model = input("Enter model: ")
    quantity = int(input("Enter quantity to sell: "))

    for bike in inventory:
        if bike['make'] == make and bike['model'] == model:
            if bike['quantity'] >= quantity:
                total_price = quantity * bike['price']
                bike['quantity'] -= quantity
                print("\n----- INVOICE -----")
                print(f"Make: {make}")
                print(f"Model: {model}")
                print(f"Quantity: {quantity}")
                print(f"Unit Price: LKR {bike['price']}")
                print(f"Total Price: LKR {total_price}")
                print("-------------------\n")
                return
            else:
                print("Not enough stock.")
                return
    print("Motorcycle not found.")

# Function to display the inventory
def display_inventory():
    print("\n----- INVENTORY REPORT -----")
    for bike in inventory:
        print(f"Make: {bike['make']}, Model: {bike['model']}, Quantity: {bike['quantity']}")
    print("-----------------------------\n")

# CLI Menu Function
def main_menu():
    while True:
        print("\n=== Madhumal Motors System ===")
        print("1. Add Motorcycle")
        print("2. Search Price")
        print("3. Sell Motorcycle (Create Invoice)")
        print("4. Display Inventory Report")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            add_motorcycle()
        elif choice == '2':
            search_price()
        elif choice == '3':
            sell_motorcycle()
        elif choice == '4':
            display_inventory()
        elif choice == '5':
            print("Exiting system...")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the Menu
main_menu()
```

```python
# Madhumal Motors CLI-Based Inventory and Sales Management System 02

# Inventory List
inventory = []

# Helper function to normalize make and model input
def normalize_input(text):
    return text.strip().lower()

# Helper function to validate make (alphabet only)
def is_valid_make(make):
    return make.isalpha()

# Helper function to validate model (numeric only)
def is_valid_model(model):
    return model.isdigit()

# Function to get existing makes from inventory
def get_existing_makes():
    return list(set(bike['make'] for bike in inventory))

# Function to add a motorcycle
def add_motorcycle():
    existing_makes = get_existing_makes()
    if existing_makes:
        print("Existing makes in system:", ', '.join(make.title() for make in existing_makes))

    make = normalize_input(input("Enter make (letters only): "))
    if not is_valid_make(make):
        print("Invalid make. Please enter letters only (e.g., Honda).")
        return

    model = normalize_input(input("Enter model (numbers only): "))
    if not is_valid_model(model):
        print("Invalid model. Please enter numbers only (e.g., 350).")
        return

    try:
        quantity = int(input("Enter quantity: "))
        price = float(input("Enter price: "))
    except ValueError:
        print("Invalid input for quantity or price. Please enter numbers.")
        return

    for bike in inventory:
        if bike['make'] == make and bike['model'] == model:
            bike['quantity'] += quantity
            print("Updated existing bike quantity.")
            return

    inventory.append({
        'make': make,
        'model': model,
        'quantity': quantity,
        'price': price
    })
    print("Motorcycle added successfully.")

# Function to search for a motorcycle price
def search_price():
    make = normalize_input(input("Enter make: "))
    model = normalize_input(input("Enter model: "))

    for bike in inventory:
        if bike['make'] == make and bike['model'] == model:
            print(f"The price of {make.title()} {model} is LKR {bike['price']}")
            return
    print("Motorcycle not found.")

# Function to sell motorcycle and generate invoice
def sell_motorcycle():
    make = normalize_input(input("Enter make: "))
    model = normalize_input(input("Enter model: "))

    try:
        quantity = int(input("Enter quantity to sell: "))
    except ValueError:
        print("Invalid quantity. Please enter a number.")
        return

    for bike in inventory:
        if bike['make'] == make and bike['model'] == model:
            if bike['quantity'] >= quantity:
                total_price = quantity * bike['price']
                bike['quantity'] -= quantity
                print("\n----- INVOICE -----")
                print(f"Make: {make.title()}")
                print(f"Model: {model}")
                print(f"Quantity: {quantity}")
                print(f"Unit Price: LKR {bike['price']}")
                print(f"Total Price: LKR {total_price}")
                print("-------------------\n")
                return
            else:
                print("Not enough stock.")
                return
    print("Motorcycle not found.")

# Function to display the inventory
def display_inventory():
    print("\n----- INVENTORY REPORT -----")
    for bike in inventory:
        print(f"Make: {bike['make'].title()}, Model: {bike['model']}, Quantity: {bike['quantity']}")
    print("-----------------------------\n")

# CLI Menu Function
def main_menu():
    while True:
        print("\n=== Madhumal Motors System ===")
        print("1. Add Motorcycle")
        print("2. Search Price")
        print("3. Sell Motorcycle (Create Invoice)")
        print("4. Display Inventory Report")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            add_motorcycle()
        elif choice == '2':
            search_price()
        elif choice == '3':
            sell_motorcycle()
        elif choice == '4':
            display_inventory()
        elif choice == '5':
            print("Exiting system...")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the Menu
main_menu()
```
